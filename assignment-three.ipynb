```python
# Define pre- and post-treatment periods and select relevant columns
df['date'] = pd.to_datetime(df['date'])
pre_event_date = '2021-01-06'

# Mark whether data is pre- or post-event(0 = pre-event, 1 = post-event)
df['post_event'] = (df['date'] > pre_event_date).astype(int)

# Create a summary of the data for DiD purposes
df_did = df[['fake_merged_rt', 'group', 'post_event']].copy()

# Group by group and post_event and calculate the mean of fake_merged_rt
did_summary = df_did.groupby(['group', 'post_event']).agg({'fake_merged_rt': 'mean'}).reset_index()

# Display the summary of the data
did_summary
```

Key Groups:

Group A: Misinformation sharers following Trump but no other deplatformed users. Average retweets slightly increased from 63.46 (pre-event) to 69.61 (post-event). Subgroups (high, moderate, low activity) show mixed behavior.

Group B: Misinformation sharers following Trump or any deplatformed user. Average retweets dropped significantly from 2224.55 to 1095.99. All subgroups show major reductions.

Group D: Misinformation sharers following at least four deplatformed users, including Trump. Average retweets fell from 1965.13 to 881.60. Subgroups also show reductions.

Group F: Misinformation sharers not following any deplatformed user. Average retweets decreased from 244.99 to 141.05, with reductions across subgroups.

Misinformation Sharers (fns): Users who shared at least one misinformation URL. Average retweets decreased from 3189.66 to 1242.06.

Deplatformed Users: Suspended users saw a near-complete drop in average retweets from 722.07 to 5.57.

QAnon and Antivax:
    QAnon: Retweets dropped from 1184.34 to 216.55.
    Antivax: Retweets dropped from 967.96 to 198.43.

```python
import matplotlib.pyplot as plt
import seaborn as sns

did_summary['post_event'] = did_summary['post_event'].astype(str)

plt.figure(figsize=(25, 10))
sns.barplot(x='group', y='fake_merged_rt', hue='post_event', data=did_summary, palette='coolwarm')

# Add labels and title
plt.title('Mean Misinformation Retweets Before and After the Event by Group', fontsize=20)
plt.xlabel('Group', fontsize=12)
plt.ylabel('Mean Misinformation Retweets', fontsize=12)
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

# **Section 3** Replication of Main DiD Results
In this section, you will perform at least one Difference in Differences analysis with the goal of conceptually replicating the key DiD analysis that McCabe et al performed to support their primary conclusion.  

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

DiD for Deplatformed Users

```python
df['treatment'] = (df['group'] == 'suspended').astype(int)

did_model_suspended = smf.ols('fake_merged_rt ~ treatment + post_event + treatment*post_event', data=df).fit()
print("Deplatformed Users DiD Results:")
print(did_model_suspended.summary())
```

DiD for Group B (Deplatformed Followers, including Trump followers)

```python
df['treatment'] = (df['group'] == 'B').astype(int)

did_model_B = smf.ols('fake_merged_rt ~ treatment + post_event + treatment*post_event', data=df).fit()
print("\nGroup B DiD Results:")
print(did_model_B.summary())
```

DiD for Group D (Four+ Deplatformed Followers, including Trump)

```python
df['treatment'] = (df['group'] == 'D').astype(int)

did_model_D = smf.ols('fake_merged_rt ~ treatment + post_event + treatment*post_event', data=df).fit()
print("\nGroup D DiD Results:")
print(did_model_D.summary())
```

DiD for High Activity Users (ha)

```python
df['treatment'] = (df['group'] == 'ha').astype(int)

did_model_ha = smf.ols('fake_merged_rt ~ treatment + post_event + treatment*post_event', data=df).fit()
print("\nHigh Activity Users DiD Results:")
print(did_model_ha.summary())
```

```python
df['treatment'] = (df['group'] == 'ma').astype(int)

did_model_ma = smf.ols('fake_merged_rt ~ treatment + post_event + treatment*post_event', data=df).fit()
print("\nModerate Activity Users DiD Results:")
print(did_model_ma.summary())
```

Question: Synchronous Exit of Right-wing (Conservative) Users

We hypothesize that some users, who were not deplatformed, might have voluntarily left Twitter after January 6th. If a large number of these users exited Twitter, this could lead to a drop in misinformation retweets, but the effect would not be due to deplatforming.

We can use both liberal and conservative non-misinformation data and leverage them as comparison groups. This will help us identify whether the drop in misinformation activity is specific to right-wing/conservative users or part of a broader trend across all users.

Conservative Non-Misinformation Posts can act as a proxy for right-wing users' overall activity that is not misinformation-related. If conservative non-misinformation posts drop alongside misinformation posts, it could indicate that right-wing users voluntarily left the platform.

Liberal Non-Misinformation Posts can serve as a control group because we expect that liberal users are less likely to have left the platform voluntarily. If liberal non-misinformation posts remain steady, while conservative posts drop, it strengthens the argument that right-wing users may have voluntarily exited Twitter.

If we see a drop in both conservative non-misinformation posts and misinformation from QAnon followers, while liberal activity stays stable, this points to voluntary exit by conservative users.
If both liberal and conservative activity remains stable, but misinformation activity decreases, the effect of deplatforming is likely responsible for the reduction in misinformation. 

```python
# Filter for Misinformation Sharers (MS) group
ms_df = df[df['group'] == 'fns']

# Analyze Conservative Non-Misinformation Posts/Retweets
plt.figure(figsize=(10, 6))
sns.lineplot(x='date', y='not_fake_conservative', data=ms_df, marker='o', label='Conservative Non-Misinformation Posts')
plt.axvline(pd.to_datetime(pre_event_date), color='red', linestyle='--', label='Jan 6, 2021')
plt.title('Conservative Non-Misinformation Activity from Misinformation Sharers (MS) Over Time')
plt.ylabel('Number of Posts (Conservative Non-Misinformation)')
plt.xlabel('Date')
plt.legend()
plt.tight_layout()
plt.show()

# Analyze Liberal Non-Misinformation Posts/Retweets
plt.figure(figsize=(10, 6))
sns.lineplot(x='date', y='not_fake_liberal', data=ms_df, marker='o', label='Liberal Non-Misinformation Posts')
plt.axvline(pd.to_datetime(pre_event_date), color='red', linestyle='--', label='Jan 6, 2021')
plt.title('Liberal Non-Misinformation Activity from Misinformation Sharers (MS) Over Time')
plt.ylabel('Number of Posts (Liberal Non-Misinformation)')
plt.xlabel('Date')
plt.legend()
plt.tight_layout()
plt.show()

# Calculate pre-post event averages for Conservative Non-Misinformation Activity
pre_event_conservative = ms_df[ms_df['post_event'] == 0]['not_fake_conservative'].mean()
post_event_conservative = ms_df[ms_df['post_event'] == 1]['not_fake_conservative'].mean()

# Calculate pre-post event averages for Liberal Non-Misinformation Activity
pre_event_liberal = ms_df[ms_df['post_event'] == 0]['not_fake_liberal'].mean()
post_event_liberal = ms_df[ms_df['post_event'] == 1]['not_fake_liberal'].mean()

# Print the results
print(f'Average Conservative Non-Misinformation Posts from Misinformation Sharers (MS):')
print(f'Pre-Event (Before Jan 6, 2021): {pre_event_conservative:.2f}')
print(f'Post-Event (After Jan 6, 2021): {post_event_conservative:.2f}')

print(f'\nAverage Liberal Non-Misinformation Posts from Misinformation Sharers (MS):')
print(f'Pre-Event (Before Jan 6, 2021): {pre_event_liberal:.2f}')
print(f'Post-Event (After Jan 6, 2021): {post_event_liberal:.2f}')

# Compare Conservative Initiation and Retweet Activity Separately
pre_event_conservative_initiation = ms_df[ms_df['post_event'] == 0]['not_fake_conservative_initiation'].mean()
post_event_conservative_initiation = ms_df[ms_df['post_event'] == 1]['not_fake_conservative_initiation'].mean()

pre_event_conservative_retweets = ms_df[ms_df['post_event'] == 0]['not_fake_conservative_rt'].mean()
post_event_conservative_retweets = ms_df[ms_df['post_event'] == 1]['not_fake_conservative_rt'].mean()

# Print results for Conservative Initiation and Retweets
print(f'Average Conservative Non-Misinformation Initiation:')
print(f'Pre-Event: {pre_event_conservative_initiation:.2f}')
print(f'Post-Event: {post_event_conservative_initiation:.2f}')

print(f'Average Conservative Non-Misinformation Retweets:')
print(f'Pre-Event: {pre_event_conservative_retweets:.2f}')
print(f'Post-Event: {post_event_conservative_retweets:.2f}')
```

As we can see, right-wing content creation and engagement decreased notably after January 6. There is smaller decline on liberal activity, pointing to a more conservative-specific reduction.

 There is a sharp drop in Conservative Non-Misinformation retweets indicates a significant reduction in engagement, implying fewer users interacting with content post-event. Conservative non-misinformation initiations also dropped; however, the decline (772 to 527) was smaller than for retweets, indicating slowed but continued content creation.

Overall:

    Conservative activity dropped sharply, particularly in retweets, indicating potential voluntary exit or reduced engagement by right-wing users.
    Liberal activity showed a smaller decline, supporting that the conservative reduction was likely specific to right-wing users rather than a platform-wide trend.

This suggests the decline in misinformation could be due to a mix of deplatforming and voluntary exit by right-wing users.

According to the codebook, misinformation tweets were classified using sources like the Grinberg et al. (2019) and Newsguard lists, identifying tweets that contained URLs or content from known misinformation sites. While this approach relies on widely recognized datasets, making it generally reliable, its fairness depends on how comprehensive these lists are. Some misinformation might have been excluded if it wasn't included in these specific sources.

The classification of the current dataset is clear, fair, and useful, as it accounts for different levels of engagement with deplatformed accounts. However, it assumes that following an account means aligning with or participating in misinformation, which isn‚Äôt always the case. Some users may follow deplatformed accounts simply to track their activities or oppose their views. A more accurate approach would classify followers based on their engagement behaviors, such as retweeting or liking misinformation, rather than just following.

According to the codebook, users were classified as misinformation sharers if they shared at least one URL flagged as misinformation in sources like the Grinberg or Newsguard lists, without considering their additional comments on the retweet. This approach may be too lenient, as it labels someone as a misinformation sharer based on a single link. A stricter classification could require users to share multiple pieces of (maybe similar) misinformation over time or consider the comments along with their retweet. Additionally, people sometimes repost content to criticize it, not to support it, which isn‚Äôt captured by this method.

The DiD analysis suggests that deplatforming did reduce misinformation for most of the user groups on Twitter as is demonstrated in Section 3, though the reduction was likely driven by a combination of deplatforming and voluntary exit of right-wing users as is explained in Section 4. 

The conclusions from McCabe et al. (2024) seem valid, as deplatforming likely played a key role in reducing misinformation. However, the effect of voluntary exit shouldn‚Äôt be ignored. Many conservative users probably left Twitter for other platforms, and it's better to monitor the user activities before and after Jan 6th on other right-wing related platforms to draw a clearer conclusion whether the conservatives are moveing to other platforms.
